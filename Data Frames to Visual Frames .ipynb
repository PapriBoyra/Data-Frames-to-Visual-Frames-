# **Line plots**

Line plots are used to visualise numeric variables, particularly if one of the variables is sequential in nature. We use a different dataset containing sequential data for this section. Line plots are useful for seeing trends in time-series data.

Plotting the number of passengers in the given airline data over time

```python
df_ap.head()
```

```python
plt.figure(figsize = (10, 4))
plt.plot(df_ap['Month'],df_ap['Passengers'])
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.ylabel('Passengers');
```

The range() function in Matplotlib is used to set the axis range.

```python
plt.plot(df_ap['Month'], df_ap['Passengers'])
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.xticks(range(23, 144, 24));
```

The plt.xticks in Matplotlib is used to customise the tick marks on the x-axis of a plot. It allows to control the placement, labels, rotation, and appearance of these ticks.

Adding markers to the line plot

```python
plt.plot(df_ap['Month'], df_ap['Passengers'],markerfacecolor='orange', markeredgecolor='black', marker = 'p')
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.xticks(range(23, 144, 24));
```

The figsize parameter of plt.figure() is used to set size of the output plot.

```python
plt.figure(figsize = (10, 4))
plt.plot(df_ap['Month'], df_ap['Passengers'], marker = 'o',markerfacecolor='red', markeredgecolor='black', markersize = 5)
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.xticks(range(23, 144, 24));
```

```python
plt.figure(figsize = (10, 4))
plt.plot(df_ap['Month'], df_ap['Passengers'], marker = 'p',markerfacecolor='orange', markeredgecolor='black', markersize = 5)
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.xticks(range(23, 144, 24))
plt.xlim((0, 150))
plt.ylim((0, 750));
```

plt.ylim in Matplotlib is used to set or retrieve the limits of the y-axis in a plot. It allows you to control the range of values that are displayed on the y-axis. The same happens for plt.xlim. In the above plot, the x-axis ranges from 0 to 150 months and the y-axis ranges from 0 to 750 passengers.

A histogram is a chart which plots the distribution of a numeric variable by representing it as a series of bars.

```python

plt.hist(df['age'])
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.title('Distribution of ages of apeople');
```

```python
plt.hist(df['age'], bins = 5)
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.title('Distribution of ages of people');
```

Let's now try to draw different histograms based on gender.

```python
male = df[df['sex'] == 1]
female = df[df['sex'] == 0]
```

```python
plt.hist(male['age'])
plt.hist(female['age'])
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.title('Distribution of ages of people based on gender');
```

In the above histogram, it is difficult to understand which colour represents which gender. Hence, we add a legend.

```python
plt.hist(male['age'], label = 'male')
plt.hist(female['age'], label = 'female')
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.title('Distribution of ages of people based on gender')
plt.legend();
```

An alternate way of representing the same information is by using a stacked histogram, or by plotting the histograms in different subplots.

```python
plt.hist([male['age'], female['age']], alpha = 0.5, label = ['male', 'female'], stacked = True)
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.title('Distribution of ages of people based on gender')
plt.legend();
```

The subplot() function below is a matplotlib function used for creating subplots.

```python
plt.figure(figsize = (8, 4))
plt.subplot(1, 2, 1)
plt.hist(male['age'], bins=20, color='blue')
plt.title('Histogram of ages of Men')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
plt.hist(female['age'], bins=20, color='green')
plt.title('Histogram of ages of Women')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xticks(rotation=45)


plt.yticks(range(0,35,5))
plt.suptitle('Histograms of ages of Men and Women')
plt.tight_layout()
plt.show()
```

Multiple categorical variables using stacked and grouped bar plots.

```python
df_churn['PaperlessBilling'].head()
```

```python
df_churn['PaymentMethod'].head()
```

```python
billing_payment_method_crosstable = pd.crosstab(df_churn['PaperlessBilling'], df_churn['PaymentMethod'])
billing_payment_method_crosstable
```

```python
payment_categories = billing_payment_method_crosstable.columns
```

```python
x = np.arange(len(payment_categories))
plt.bar(x, billing_payment_method_crosstable.loc['Yes'], label='Yes')
plt.bar(x, billing_payment_method_crosstable.loc['No'], label='No')
plt.xlabel('Payment methods')
plt.ylabel('Count')
plt.xticks(x, labels=payment_categories)
plt.title('Distribution of number of payments over billing type and payment method')
plt.legend()
plt.show()
```

The last bar graph is not a correct visualisation as the 'No' bar completely overlaps the 'Yes' bar. A grouped bar plot is more suitable.

Changing the width of bars in the bar graph

```python
x = np.arange(len(payment_categories))
plt.bar(x, billing_payment_method_crosstable.loc['Yes'], label='Yes',width=0.5)
plt.bar(x, billing_payment_method_crosstable.loc['No'], label='No',width=0.5)
plt.xlabel('Payment methods')
plt.ylabel('Count')
plt.xticks(x, labels=payment_categories)
plt.title('Distribution of number of payments over billing type and payment method')
plt.legend()
plt.show()
```

Changing to a grouped bar plot

```python
x = np.arange(len(payment_categories))
plt.bar(x-0.2, billing_payment_method_crosstable.loc['Yes'], label='Yes',width=0.4)
plt.bar(x+0.2, billing_payment_method_crosstable.loc['No'], label='No',width=0.4)
plt.xlabel('Payment methods')
plt.ylabel('Count')
plt.xticks(x, labels=payment_categories)
plt.title('Distribution of number of payments over billing type and payment method')
plt.legend()
plt.show()
```

# **Box plots**

A box plot, also known as a box-and-whisker plot, is a statistical visualization tool that provides a summary of the distribution of a dataset. It displays the central tendency, spread, and identifies potential outliers. Here's a breakdown of the components of a box plot:


Box (Interquartile range or IQR): The box represents the interquartile range, which is the range between the first quartile (Q1) and the third quartile (Q3). The length of the box signifies the spread of the middle  50 % of the data.


Line inside the box (Median): A line inside the box represents the median (Q2), which is the middle value of the dataset when it is sorted in ascending order (or descending order).


Whiskers: Whiskers extend from the box to the minimum and maximum values within a specified range (usually  1.5  times the IQR). They show the spread of the entire dataset.


Outliers: Individual points beyond the whiskers are considered outliers. They are potential data points that fall significantly outside the overall distribution.

Generating a boxplot that visualises the distribution of the 'MonthlyCharges' column from the df_churn DataFrame.

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df_churn['MonthlyCharges'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))

plt.title('Boxplot of Monthly Charges', fontsize=16)
plt.xlabel('Monthly Charges', fontsize=12)
#We can customize the yticks text using the code below
plt.yticks([1], ['Boxplot'])
plt.show()
```

# **Sub plot**

A layout mechanism that allows you to display multiple plots (which could include boxplots, histograms, line plots, etc.) together in a single figure by arranging them in a grid.

Creating a subplot with two different plots side by side

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

axs[0].hist(df_churn['tenure'], bins=20, color='skyblue', edgecolor='black')
axs[0].set_title('Distribution of Tenure')
axs[0].set_xlabel('Tenure')
axs[0].set_ylabel('Frequency')

axs[1].boxplot(df_churn['MonthlyCharges'], vert=False, patch_artist=True)
axs[1].set_title('Boxplot of Monthly Charges')
axs[1].set_xlabel('Monthly Charges')

plt.tight_layout()
plt.show()
```

A histogram of the tenure column and a boxplot of the 'MonthlyCharges' column from the df_churn DataFrame are shown above.
